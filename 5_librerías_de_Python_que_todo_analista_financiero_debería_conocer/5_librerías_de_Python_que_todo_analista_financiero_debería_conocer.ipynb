### C贸digo de pandas

```python
import pandas as pd #Importante importar la paqueria a utilizar

# Simulaci贸n de precios de una acci贸n
data = {
    "Fecha": pd.date_range(start="2025-01-01", periods=5, freq="D"),
    "Precio": [100, 102, 101, 105, 107]
}

df = pd.DataFrame(data)
print(f"{df}\n")

# Calcular rendimientos simples
df["Rendimiento"] = df["Precio"].pct_change() # Funcion para Calcula el cambio fraccional de la fila inmediatamente anterior

print(df)
```

### Introducci贸n

### Las operaciones financieras suelen requerir mucho c谩lculo num茅rico: medias, desviaciones, tasas compuestas. Ah铆 entra NumPy.

### Fundamento

### Permite trabajar con vectores y matrices, optimizando c谩lculos repetitivos.

### Ejemplo: Rendimiento compuesto

$$ R_c = \left( \prod_{t=1}^{n} (1 + r_t) \right)^{-1} $$

### C贸digo de Numpy

```python
import numpy as np #Importante importar la paqueria a utilizar

rendimientos = np.array([0.02, -0.01, 0.04, 0.03])  # 4 periodos
rend_compuesto = np.prod(1 + rendimientos) - 1
print("Rendimiento compuesto:", round(rend_compuesto, 4))
```

En este c贸digo, usamos la biblioteca numpy para calcular el rendimiento compuesto total a partir de una serie de rendimientos simples de varios periodos.

Hacemos esto para conocer la ganancia o p茅rdida acumulada de una inversi贸n, considerando que las ganancias de un periodo se reinvierten y generan ganancias en el siguiente.

Finalmente, el c贸digo muestra en una sola l铆nea el resultado de ese c谩lculo total, redondeado a 4 decimales.

### 5锔 Statsmodels: Entendiendo relaciones con econometr铆a
### Introducci贸n

### En an谩lisis financiero no solo observamos, tambi茅n queremos explicar y predecir. Statsmodels permite hacer regresiones y an谩lisis estad铆stico.

### Fundamento

### Ejemplo de regresi贸n lineal:

### $$R_{\text{stock}} = \alpha + \beta \cdot R_{\text{market}} + \epsilon$$
Donde:
*   ####  = rendimiento independiente del mercado (alpha).
*   ####  = sensibilidad al mercado.
*   ####  = error aleatorio.

### C贸digo de Statsmodels

```python
import statsmodels.api as sm

# Datos simulados
rend_mercado = np.random.normal(0.01, 0.02, 100)
rend_accion = 0.002 + 1.2 * rend_mercado + np.random.normal(0, 0.01, 100)

X = sm.add_constant(rend_mercado)
modelo = sm.OLS(rend_accion, X).fit()
print(modelo.summary())
```

Interpretaci贸n

Si 尾 > 1: la acci贸n es m谩s vol谩til que el mercado.

Si 尾 < 1: la acci贸n es m谩s defensiva.

Si 伪 > 0: la acci贸n genera exceso de rendimiento independiente del mercado.

En este c贸digo, primero simulamos datos para los rendimientos de un mercado y de una acci贸n, creando una relaci贸n lineal entre ellos. Despu茅s, usamos la biblioteca statsmodels para aplicar un modelo de regresi贸n lineal (conocido como M铆nimos Cuadrados Ordinarios - OLS).

Hacemos esto para medir estad铆sticamente la relaci贸n entre la acci贸n y el mercado. El objetivo es estimar el:

Alpha (伪): El intercepto, que representa el rendimiento de la acci贸n que no es explicado por los movimientos del mercado.

Beta (尾): La pendiente, que mide la sensibilidad o volatilidad de la acci贸n en relaci贸n con los movimientos del mercado.

Finalmente, el c贸digo muestra una tabla de resumen estad铆stico con los resultados completos de la regresi贸n. Esta tabla incluye valores clave como el R-cuadrado (qu茅 tan bien el modelo explica los datos) y los coeficientes calculados para la constante (Alpha) y la variable del mercado (Beta).
